```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = ["The movie was great and I really enjoyed it.",
          "The movie was boring and I didn't enjoy it at all."]

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(corpus)

# This will print the feature names (words)
print(vectorizer.get_feature_names_out())
# This will print the TF-IDF matrix (sparse matrix)
print(X.toarray())
```
